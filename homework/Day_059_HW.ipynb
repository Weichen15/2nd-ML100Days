### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets

digits = datasets.load_digits(n_class=3)
X1 = digits.data
X2 = digits.data
y = digits.target
X1, y
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


from sklearn import decomposition
#PCA的模塊，在decompsition中

np.random.seed(0)
%matplotlib inline
```

```python
fig = plt.figure(1, figsize=(4, 3))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
plt.cla()


for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X1[y == label, 0].mean(),
              X1[y == label, 1].mean() + 1.5,
              X1[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
y1 = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X1[:, 0], X1[:, 1], X1[:, 2], c=y1, cmap=plt.cm.nipy_spectral,
           edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
centers = [[1, 1], [-1, -1], [1, -1]]
pca = decomposition.PCA(n_components=3)
```

```python
pca.fit(X2)
X2 = pca.transform(X2)

fig = plt.figure(1, figsize=(4, 3))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
plt.cla()


for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X2[y == label, 0].mean(),
              X2[y == label, 1].mean() + 1.5,
              X2[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
y2 = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X2[:, 0], X2[:, 1], X2[:, 2], c=y2, cmap=plt.cm.nipy_spectral,
           edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
